### 1. LOAD
   - Concatenated 4 class CSVs into df_students
   - Loaded claustro.csv into df_faculty

```python
# we load all of the class CSVs and stack them into just one dataframe

df_students = pd.concat([
    pd.read_csv("data/clase_1.csv", sep = ";"),
    pd.read_csv("data/clase_2.csv", sep = ";"),
    pd.read_csv("data/clase_3.csv", sep = ";"),
    pd.read_csv("data/clase_4.csv", sep = ";")

], ignore_index=True)
```

```python
# load faculty CSV separately (different structure)

df_faculty = pd.read_csv("data/claustro.csv", sep=";")
```

### 2. CLEAN
   - Fixed typo (FullSatck → FullStack)
   - Melted project columns from wide → long format (df_melted)
   - Dropped NaN rows (260 real grades remaining)

```python
# 9 columns were expected but we got 15 instead 

print(df_students.shape)
print(df_faculty.shape)
```

```python
# the 15 columns are due to the 10 projects that there are (5 for DS and 5 for FS)

print(df_students.columns.tolist())
```

```python
# 1st we see what the data looks like

df_students.head(2)
```

```python
# we correct the typo on the column named "Proyecto_FullSatck"

df_students = df_students.rename(columns = {"Proyecto_FullSatck": "Proyecto_FullStack"})
```

```python
# 2nd we see how many null values in each column of our dataframe

print(df_students.isnull().sum())
```

```python
# we create a list of the columns that we want to convert into rows 

project_columns = [
    "Proyecto_HLF", "Proyecto_EDA", "Proyecto_BBDD", "Proyecto_ML", 
    "Proyecto_Deployment", "Proyecto_WebDev", "Proyecto_FrontEnd", 
    "Proyecto_Backend", "Proyecto_React", "Proyecto_FullStack"
]
```

```python
df_melted = df_students.melt( # df_students.melt() — we're calling the melt function -> transforms the datafarme from wide to long format
    id_vars = ["Nombre", "Email", "Promoción", "Fecha_comienzo", "Campus"], # id_vars = [...] -> these are the columns we want to keep as-is
    value_vars = project_columns, # value_vars = project_columns -> these are the columns we wanto to convert into rows // instead of 10 project columns, each project becomes its own row
    var_name = "tipo_proyecto", # var_name = "tipo_proyecto" -> this is the name we give to the new column that stores the project column names (i.e, "Projecto_EDA")
    value_name = "resultado" # value_name = "resultado" -> this is the name we give to the new column that stores the values (Apto // No Apto)
)
```

```python
df_melted.head(5)
```

```python
# we need to drop the NULL values since they don't represent real grades 

df_melted = df_melted.dropna(subset = ["resultado"])
print(df_melted.shape)
```

```python
df_melted
```

### 3. LOOKUP TABLES
   - df_campus         (Madrid, Valencia...)
   - df_tipo_proyecto  (HLF, EDA, BBDD...)
   - df_promocion      (Septiembre, Febrero)
   - df_modalidad      (Presencial, Online)
   - df_vertical       (DS, FS)
   - df_rol            (Teacher, Teacher Assistant)

```python
# in our df_melted each campus is repeated multiple times
# we create a new dataframe (campus) and with "drop_duplicates" we keep only 1 campus
# we use reset_index because, for example, if "Valencia" was originally on row 515, it's index would still be 515
# drop = True: with this we're telling the system "don't keep the old index, throw it away"
# index + 1 because Postgre SQL expects IDs to start at 1

df_campus = df_melted[["Campus"]].drop_duplicates().reset_index(drop = True)
df_campus["id"] = df_campus.index + 1
print(df_campus)
```

```python
# same thing as before
# we assign each project and ID to only have one of each

df_tipo_proyecto = df_melted[["tipo_proyecto"]].drop_duplicates().reset_index(drop = True)
df_tipo_proyecto["id"] = df_tipo_proyecto.index + 1
df_tipo_proyecto = df_tipo_proyecto.rename(columns =  {"id": "tipo_proyecto_id"})
print(df_tipo_proyecto)
```

```python
# same thing as before
# we assign each promotion and ID to only have one of each

df_promocion = df_melted[["Promoción"]].drop_duplicates().reset_index(drop = True)
df_promocion["id"] = df_promocion.index + 1
print(df_promocion)
```

```python
# same thing as before
# we assign each modalidad (presencial or online) and ID to only have one of each

df_modalidad = df_faculty[["Modalidad"]].drop_duplicates().reset_index(drop = True)
df_modalidad["id"] = df_modalidad.index + 1
print(df_modalidad)
```

```python
# same thing as before
# we assign each vertical (DS or FS) and ID to only have one of each

df_vertical = df_faculty[["Vertical"]].drop_duplicates().reset_index(drop = True)
df_vertical["id"] = df_vertical.index + 1
print(df_vertical)
```

```python
# same thing as before
# we assign each role (teacher or TA) and ID to only have one of each

df_rol = df_faculty[["Rol"]].drop_duplicates().reset_index(drop = True)
df_rol["id"] = df_rol.index + 1
print(df_rol)
```

### 4. Add id to df_students
   - df_students → add estudiante_id to the dataframe

```python
df_students["id"] = df_students.index + 1
df_students = df_students.rename(columns =  {"id": "estudiante_id"})
print(df_students[["estudiante_id", "Nombre"]].head())
```

### 5. MAIN TABLES
   - df_students → merged with campus + promocion → kept Nombre, Email, campus_id, promocion_id
   - df_faculty  → merged with rol + campus + vertical → kept Nombre, rol_id, campus_id, vertical_id

```python
# we now need to create the main entity tables (studnets and faculty)
# let's start with estudiante
# how = left -> we want to keep all of the columns from df_students and match campus where possible

df_students = df_students.merge(df_campus, on = "Campus", how = "left")
print(df_students[["Nombre", "Campus", "id"]].head())
```

```python
print(df_students.columns.tolist())
print(df_campus.columns.tolist())
```

```python
# id is too generic // we'll change it to campus_id

df_students = df_students.rename(columns =  {"id": "campus_id"})
```

```python
# same as before but instead of campus we now merge with promocion

df_students = df_students.merge(df_promocion, on = "Promoción", how = "left")
print(df_students[["Nombre", "Promoción", "id"]].head())
```

```python
# id is too generic // we'll change it to promocio_id

df_students = df_students.rename(columns =  {"id": "promocion_id"})
```

```python
df_faculty.head(1)
```

```python
print(df_faculty.columns.tolist())
```

```python
# same as we did before with students
# but now we merge df_faculty with df_rol, df_campus and df_vertical
# since "id" is too generic for each merger // we modify each "id" depending on their "parent" df

df_faculty = df_faculty.merge(df_rol, on = "Rol", how = "left")
df_faculty = df_faculty.rename(columns = {"id": "rol_id"})

df_faculty = df_faculty.merge(df_campus, on = "Campus", how = "left")
df_faculty = df_faculty.rename(columns = {"id": "campus_id"})

df_faculty = df_faculty.merge(df_vertical, on = "Vertical", how = "left")
df_faculty = df_faculty.rename(columns = {"id": "vertical_id"})

df_faculty = df_faculty[["Nombre", "rol_id", "campus_id", "vertical_id"]]
print(df_faculty.head())
```

### 5. NEXT UP
   - Build df_calificacion from df_melted
   - Connect to PostgreSQL and insert everything

```python
# our df_melted currently looks like this 
# we need to replace the text values with their IDs so that PostgreSQL may read it correctly

df_melted.head(1)
```

```python
# here we merge df_melted and df_students to get "estudiante_id"
# we only merge on "Nombre" (to match on) and "estudiante_id" to keep
# everything else is uncessary

df_calificacion = df_melted.merge(df_students[["Nombre", "estudiante_id"]], on = "Nombre", how = "left")
```

```python
# same as before but in this case we merge df_califacion and df_tipo_proyecto to get ·proyecto_id"

df_calificacion = df_calificacion.merge(df_tipo_proyecto, on = "tipo_proyecto", how = "left")
```

```python
# lastly we modify our df_califacion to only keep what we need

df_calificacion = df_calificacion[["estudiante_id", "tipo_proyecto_id", "resultado"]]
```

```python
# double check to see if our df is consisted of pure IDs except the "resultado" column

print(df_calificacion.head(10))
```

```python
# one last check to make sure that we no longer have null values 

print(df_calificacion.shape)
print(df_calificacion.isnull().sum())
```
